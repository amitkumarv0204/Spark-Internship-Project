```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import decimal
```

```python
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
print("Importing Data Successfully")
df.head(5)
```

```python
df.plot(x='Hours',y='Scores',style='3')
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
df.plot.scatter(x='Hours',y='Scores')
```

```python
df.plot.bar(x='Hours',y='Scores')
```

```python
df.sort_values(["Hours"], axis=0,
                 ascending=[True])
df.head(10)
df.plot.bar(x='Hours',y='Scores')
```

```python

X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
# print(X)
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() 
regressor.fit(X_train, y_train) 
print("Training complete.")
```

```python
print(X_test) 
print("Predection of Score")
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
```

```python
hours = [[9.25]]
pred = regressor.predict(hours)
if(pred<=100):
     print(np.round(pred, 2))
else:
    print("100.00")
```

```python
from sklearn import metrics  
print('Mean Absolute Error:',np.round( 
      metrics.mean_absolute_error(y_test, y_pred),2)) 
```
